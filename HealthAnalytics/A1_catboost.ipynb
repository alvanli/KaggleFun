```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
base_dir = "C://DATA//train//49_kaggle//KaggleFun//HealthAnalytics//DATA"
df_train = pd.read_csv(os.path.join(base_dir,"pro_train.csv"))
df_test = pd.read_csv(os.path.join(base_dir,"pro_test.csv"))
df_train.head(5)
```

```python
df_y = df_train.Stay
df_train = df_train.drop(["Stay","Unnamed: 0"], axis = 1)
df_test_y = df_test.Stay
df_test = df_test.drop(["Stay","Unnamed: 0"], axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
sc=StandardScaler()
df_train = sc.fit_transform(df_train)
df_test = sc.transform(df_test)
```

```python
from catboost import Pool, CatBoostClassifier

train_dataset = Pool(data=df_train, label=df_y)

eval_dataset = Pool(data=df_test, label=df_test_y)
# initialising catboost classifier

model = CatBoostClassifier(eval_metric='Accuracy')

model.fit(train_dataset,eval_set=eval_dataset)

model.get_best_score()
```
